```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
# IFNB enhancer sequence
enName='IFNB'
en='TAGAAACTACTAAAATGTAAATGACATAGGAAAACTGAAAGGGAGAAGTGAAAGTGGGAAATTCCTCTGAATAGAGAGAGGACCATC'
```

```python
len(en)
```

```python
# Get mpra data
fn='IFNB_41587_2012_BFnbt2137_MOESM14_ESM__pvalues.tsv'
df=pd.read_csv(fn,sep='\t').set_index('nt')
df.head()
```

```python
# Get mpra data
fn='IFNB_41587_2012_BFnbt2137_MOESM14_ESM__effects-bymedian.tsv'
effDF=pd.read_csv(fn,sep='\t').set_index('nt')
effDF.head()
```

```python
# chrom
# pos
# ref
# alt
# pvalue
# effect

c2v={c:[] for c in ['chrom','pos','ref','alt','pvalue','effect']}

ntList=list('ATGC')

# iterate over each nt in the enhancer
for idx0,ref in enumerate(en):
    idx1=str(idx0+1)
    altList=[nti for nti in ntList if nti!=ref]
    
    for alt in altList:
        
        c2v['chrom'].append(enName)
        c2v['pos'].append(idx0)
        c2v['ref'].append(ref)
        c2v['alt'].append(alt)
        c2v['pvalue'].append(df.at[alt,idx1])
        c2v['effect'].append(effDF.at[alt,idx1])
        
outDF=pd.DataFrame(c2v)
outDF.head()
    
```

```python
outDF.to_csv('IFNB__for-input.tsv',index=None,sep='\t')
```
